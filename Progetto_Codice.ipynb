### 3. For each (primary) breed, determine the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed. Hint: look at the secondary_breed.

```python

# Stampo i valori raggruppati per breed_primary e breed_secondary per vedere come sono suddivisi i dati e per un check con i risultati successivi
print(df_dog.groupby(['breed_primary', 'breed_secondary'], dropna=False).size().to_markdown())
# creo una tabella, dove nelle colonne ci sono tutti i possibili breed_secondary e i relativi valori(es. Riga[Affenpinscher]-> Colonna[Chihuahua]->valore[1] ecc..)
group_breed_dogs= df_dog.groupby(['breed_primary', 'breed_secondary'], dropna=False).size().unstack(fill_value=0)
```

```python
#creo una lista con il nome del breed_primary e il valore di quanti Mixed Breed ci sono
lista_Mixed_Breed = group_breed_dogs["Mixed Breed"]
lista_Mixed_Breed
```

```python
#calcolo il totale dei cani che ci sono per ogni id
sum_total = group_breed_dogs.sum(axis=1)

for i in range(len(sum_total)):
  print("Ratio di: ", sum_total.index[i])
  print("Cani totali: ",  sum_total[i])
  #per avere il numero dei non Mixed , sottraggo al totale quelli Mixed
  num_no_Mixed_Breed = sum_total[i] - lista_Mixed_Breed[i]
  #calcolo le due percentuali, dividendo il numero dei Mixed/ totale e il numero dei non Mixed / totale
  percentuale_Mixed = (lista_Mixed_Breed[i] / sum_total[i])*100
  percentuale_no_Mixed = (num_no_Mixed_Breed / sum_total[i])*100
  print("valore percentuale Mixed: ", percentuale_Mixed)
  print("valore percentuale no Mixed: ", percentuale_no_Mixed)
  print("--------------------------------------------------------")
```

### 7. Find the dogs involved in at least 3 travels. Also list the breed of those dogs.

```python
df_dog_travel = df_dog_travel.drop(columns={"index"})
```

```python
df_nuovo_join = pd.merge(df_dog, df_dog_travel, on='id')
df_nuovo_join
```

```python
counts = df_nuovo_join['id'].value_counts()
#counts = conto quante volte appare l'id nel merge
#metodo 1:
display(counts[counts.values >= 3])

#metodo 2:
df_group_id = df_nuovo_join[["id","breed_primary","breed_secondary"]]
df_group_id_2 = df_group_id.groupby('id').first()
df2 = df_group_id.groupby('id')[['breed_primary','breed_secondary']].size().reset_index(name='number_travels')
df2= pd.merge(df2, df_group_id_2, on='id')
df2 = df2[df2["number_travels"] >=3]
df2= df2.sort_values(by='number_travels', ascending=False)
display(df2)
```
