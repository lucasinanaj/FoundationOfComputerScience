### 12.Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python

import time
from math import modf

def is_duplicate_Jaccard(row1, row2):

    if row1['breed_primary'] == row2['breed_primary'] and row1['breed_secondary'] == row2['breed_secondary'] and row1['sex'] == row2['sex']:
        description1 = row1['description'].split()
        description2 = row2['description'].split()
        shared_words =  len(list(set(description1).intersection(description2)))
        union = (len(description1) + len(description2)) - shared_words
        jaccard = float(shared_words)/union
        print("shared_words", shared_words)
        print("union ",union )
        print("jaccard",jaccard)
        print("----------------------------------")
        if jaccard >= 0.9:
          print("jaccard", jaccard)
          return True
    return False

def is_duplicate(row1, row2):
    if row1['breed_primary'] == row2['breed_primary'] and row1['breed_secondary'] == row2['breed_secondary'] and row1['sex'] == row2['sex']:
      #utilizzo set perchè mi crea un set con i valori univoci
        description1 = set(row1['description'].split())
        description2 = set(row2['description'].split())
        shared_words = len(description1.intersection(description2))*2
        sum_count = len(description1) + len(description2)
        ratio = shared_words / sum_count
        print("shared_words", shared_words)
        print("sum_count ",sum_count )
        print("ratio",ratio)
        print("----------------------------------")
        if ratio >= 0.9:
            return True
        else:
          return False
    else:
      return False


def is_duplicate_fuzz(row1, row2):
   if row1['breed_primary'] == row2['breed_primary'] and row1['breed_secondary'] == row2['breed_secondary'] and row1['sex'] == row2['sex']:
     print("Fuzz Ratio: ", fuzz.ratio(row1['description'],row2['description']))
     if fuzz.ratio(row1['description'],row2['description']) >90:
       return True
     else:
          return False
   else:
      return False


def is_duplicate_Jaccard_grouoped(row1, row2):

    #if row1['breed_primary'] == row2['breed_primary'] and row1['breed_secondary'] == row2['breed_secondary'] and row1['sex'] == row2['sex']:
    description1 = row1['description'].split()
    description2 = row2['description'].split()
    shared_words =  len(list(set(description1).intersection(description2)))
    union = (len(description1) + len(description2)) - shared_words
    jaccard = float(shared_words)/union

    if jaccard >= 0.9:
      print("jaccard", jaccard)
      return True
    return False

```

```python
df_ridotto = df_dog.sample(5000)
df_ridotto = df_ridotto.reset_index(drop = True)
df_ridotto["is_duplicate"] = False
df_ridotto["is_duplicate_Jaccard"] = False
df_ridotto["is_duplicate_fuzz"] = False
```

```python
start = time.time()
for index, row in df_ridotto.iterrows():
  for index2, row2 in df_ridotto[index + 1:].iterrows():
    if not( pd.isnull(row['description']) or pd.isnull(row2['description']) or row['description'] == '' or row2['description'] == ''):
      jaccard = is_duplicate_Jaccard(row, row2)
      if jaccard:
        df_ridotto.at[index,"is_duplicate_Jaccard"] = jaccard
        df_ridotto.at[index2,"is_duplicate_Jaccard"] = jaccard
        print("1." , row["description"])
        print("2.", row2["description"])

end = time.time()
tempo = ((end - start)/60)
fraz,intera = modf(tempo)
fraz = str("%.2f" % fraz)
fraz = fraz.replace("0.", "")
print("Tempo di esecuzione:","%.0f" % intera , "minuti e", fraz, "secondi")
```
